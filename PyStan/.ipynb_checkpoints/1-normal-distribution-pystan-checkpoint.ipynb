```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pystan import stan, StanModel

%matplotlib inline
%config InlineBackend.figure_format = "retina"
```

```python
mu_0 = 500
sigma_0 = 1.5
n = 10

y = np.random.normal(loc=mu_0, scale=sigma_0, size=[n,])
```

```python
code = """
data {
    int<lower=1> n;
    real y[n];
}
parameters {
    real mu;
    real<lower=0> sigma;
}
model {
    y ~ normal(mu, sigma);
}
"""
```

```python
data = {"n": n, 
        "y": y}

sm  = StanModel(model_code=code)
fit = sm.sampling(data=data,
                  chains=1, 
                  iter=3000, 
                  warmup=1000,
                  verbose=True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))
sns.kdeplot(fit.extract()["mu"], shade=True, ax=axes[0])
_ = axes[0].set_title("Mu", fontsize=13)
sns.kdeplot(fit.extract()["sigma"], shade=True, ax=axes[1])
_ = axes[1].set_title("Sigma", fontsize=13)
```
